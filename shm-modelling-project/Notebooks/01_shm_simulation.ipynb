## Modelling Simple Harmonic Motion with Python


### 1. Basic Simple Harmonic Motion
In this section, I model the postion of an ideal simple harmonic oscillator using: 

$$x(t) = A\cos(\omega t + \phi)$$

where: 
- $A$ is the amplitude 
- $\omega$ is the angular frequency 
- $\phi$ is the phase constant 
- $t$ is time

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1 
omega = 2 
phi = 0
```

```python
t = np.linspace(0, 10, 1000)
x = A * np.cos(omega * t + phi)
```

```python
plt.plot(t, x)
plt.title('Simple Harmonic Motion Simulation')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.show
```

#### Interpretation

The graph shows periodic oscillatory motion, where the position varies smoothly between positive and negative values over time. Since this is an ideal simple harmonic oscillator with no damping, the amplitude remains constant throughout the motion.

```python
T = 2 * np.pi / omega
print(T, 'period of oscillation')
```

The period of the oscillator is approximately 3.14 seconds. This means the system completes one full oscillation every 3.14 seconds. Since the angular frequency was set to $\omega = 2$, the period is $T = \frac{2\pi}{\omega} = \pi$.

```python
omega_values = [1, 2, 4]
```

```python
for omega in omega_values:
    x = A * np.cos(omega * t + phi)
    plt.plot(t, x, label=f"omega = {omega}")
    plt.title("Effect of Angular Frequency on Simple Harmonic Motion")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.legend()
plt.show()
```

### Interpretation

The plot shows that increasing angular frequency increases the rate of oscillation. For $\omega = 1$, the oscillator completes fewer cycles over the 10-second interval, while for $\omega = 4$, it completes many more cycles in the same amount of time.

This happens because the period is inversely related to angular frequency:

$$
T = \frac{2\pi}{\omega}
$$

As $\omega$ increases, $T$ decreases, so each oscillation takes less time.

omega = 1 → T = $2\pi$,
omega = 2 → T = $\pi$,
omega = 4 → T = $0.5\pi$

## 2. Velocity and Acceleration in Simple Harmonic Motion

For a simple harmonic oscillator:

$$x(t) = A\cos(\omega t + \phi)$$

The velocity is the time derivative of position:

$$v(t) = -A\omega \sin(\omega t + \phi)$$

The acceleration is the time derivative of velocity:

$$a(t) = -A\omega^2 \cos(\omega t + \phi)$$

This means the acceleration is proportional to displacement but points in the opposite direction:

$$a(t) = -\omega^2 x(t)$$

```python
A = 1
omega = 2
phi = 0

t = np.linspace(0, 10, 1000)

x = A * np.cos(omega * t + phi)
v = -A * omega * np.sin(omega * t + phi)
a = -A * omega**2 * np.cos(omega * t + phi)
```

```python
plt.plot(t, x)
plt.title("Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.show()
```

```python
plt.plot(t, v)
plt.title("Velocity vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Velocity")
plt.show()
```

```python
plt.plot(t, a)
plt.title("Acceleration vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration")
plt.show()
```

### Interpretation

The position, velocity, and acceleration all oscillate periodically. The velocity is phase-shifted relative to the position, reaching its maximum magnitude when the oscillator passes through equilibrium. The acceleration is proportional to displacement but points in the opposite direction, as shown by the relationship $a(t) = -\omega^2 x(t)$.

This restoring acceleration is what causes the system to continue oscillating.

## 3. Mass-Spring Oscillator

For a mass attached to a spring, the angular frequency is determined by:

$$
\omega = \sqrt{\frac{k}{m}}
$$

where:
- $k$ is the spring constant
- $m$ is the mass

The period is:

$$
T = 2\pi\sqrt{\frac{m}{k}}
$$

This means that increasing the mass makes the oscillator slower, while increasing the spring constant makes the oscillator faster.

```python
m = 1.0
k = 4.0

omega = np.sqrt(k / m)
T = 2 * np.pi * np.sqrt(m / k)

print("Angular frequency:", omega)
print("Period:", T)
```

```python
A = 1
phi = 0
t = np.linspace(0, 10, 1000)

x = A * np.cos(omega * t + phi)

plt.plot(t, x)
plt.title("Mass-Spring Oscillator: Position vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.show()
```

### Interpretation

For a mass-spring system with \(m = 1.0\) and \(k = 4.0\), the angular frequency is \(\omega = 2.0\), giving a period of approximately \(3.14\) seconds. This matches the earlier simple harmonic motion example where \(\omega = 2\).

The motion remains periodic because the spring provides a restoring force proportional to displacement.

## 4. Effect of Mass on Oscillation

For a mass-spring oscillator,

$$
\omega = \sqrt{\frac{k}{m}}
$$

so increasing the mass decreases the angular frequency and increases the period. This means heavier masses oscillate more slowly when the spring constant is fixed.

```python
k = 4.0
A = 1
phi = 0
t = np.linspace(0, 10, 1000)

m_values = [0.5, 1.0, 2.0]

for m in m_values:
    omega = np.sqrt(k / m)
    x = A * np.cos(omega * t + phi)
    plt.plot(t, x, label=f"m = {m}, ω = {omega:.2f}")

plt.title("Effect of Mass on Mass-Spring Oscillation")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.legend()
plt.show()
```

### Interpretation

As mass increases, the oscillations become slower. This happens because the angular frequency decreases according to $\omega = \sqrt{k/m}$. For a fixed spring constant, a larger mass has greater inertia, so it takes longer for the system to complete each oscillation.

## 6. Effect of Spring Constant on Oscillation

For a mass-spring oscillator,

$$
\omega = \sqrt{\frac{k}{m}}
$$

so increasing the spring constant increases the angular frequency and decreases the period. This means a stiffer spring causes the system to oscillate faster when the mass is fixed.

```python
m = 1.0
A = 1
phi = 0
t = np.linspace(0, 10, 1000)

k_values = [1.0, 4.0, 9.0]

for k in k_values:
    omega = np.sqrt(k / m)
    x = A * np.cos(omega * t + phi)
    plt.plot(t, x, label=f"k = {k}, ω = {omega:.2f}")

plt.title("Effect of Spring Constant on Mass-Spring Oscillation")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.legend()
plt.show()
```

### Interpretation

As the spring constant increases, the oscillations become faster. This happens because the angular frequency increases according to $\omega = \sqrt{k/m}$. For a fixed mass, a larger spring constant means a stronger restoring force, so the system returns toward equilibrium more quickly and completes each oscillation in less time.


## 7. Energy in Simple Harmonic Motion

For a mass-spring oscillator, the total mechanical energy is the sum of kinetic energy and potential energy.

The kinetic energy is:

$$
K = \frac{1}{2}mv^2
$$

The spring potential energy is:

$$
U = \frac{1}{2}kx^2
$$

For an ideal undamped oscillator, energy continuously transfers between kinetic and potential energy, but the total energy remains constant.

```python
m = 1.0
k = 4.0
A = 1.0
phi = 0

omega = np.sqrt(k / m)
t = np.linspace(0, 10, 1000)

x = A * np.cos(omega * t + phi)
v = -A * omega * np.sin(omega * t + phi)

kinetic_energy = 0.5 * m * v**2
potential_energy = 0.5 * k * x**2
total_energy = kinetic_energy + potential_energy
```

```python
plt.plot(t, kinetic_energy, label="Kinetic Energy")
plt.plot(t, potential_energy, label="Potential Energy")
plt.plot(t, total_energy, label="Total Energy")

plt.title("Energy in Simple Harmonic Motion")
plt.xlabel("Time (s)")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

### Interpretation

The kinetic and potential energies oscillate over time. When the oscillator is at maximum displacement, the potential energy is highest and the kinetic energy is lowest. When the oscillator passes through equilibrium, the kinetic energy is highest and the potential energy is lowest.

The total energy remains constant throughout the motion, showing conservation of mechanical energy in an ideal undamped oscillator.

## 8. Damped Simple Harmonic Motion

In real systems, oscillations often decrease in amplitude over time because energy is lost due to friction, air resistance, or other dissipative forces.

A simple damped oscillation can be modeled as:

$$
x(t) = A e^{-\gamma t}\cos(\omega t + \phi)
$$

where:
- $A$ is the initial amplitude
- $\gamma$ is the damping coefficient
- $\omega$ is the angular frequency
- $\phi$ is the phase constant
- $t$ is time

As $\gamma$ increases, the amplitude decays more quickly.

```python
A = 1.0
omega = 2.0
phi = 0
gamma = 0.2

t = np.linspace(0, 20, 2000)

x_damped = A * np.exp(-gamma * t) * np.cos(omega * t + phi)

plt.plot(t, x_damped)

plt.title("Damped Simple Harmonic Motion")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.show()
```

### Interpretation

The damped oscillator continues to oscillate, but its amplitude decreases over time. This happens because the exponential factor \(e^{-\gamma t}\) reduces the size of the oscillations as time increases.

Unlike the ideal simple harmonic oscillator, the damped oscillator does not conserve mechanical energy because energy is gradually lost from the system.

## 9. Effect of Damping Coefficient on Oscillation

The damping coefficient $\gamma\$ controls how quickly the amplitude decreases over time.

For damped simple harmonic motion:

$$
x(t) = A e^{-\gamma t}\cos(\omega t + \phi)
$$

A larger value of $\gamma$ causes the oscillations to decay more quickly, while $\gamma = 0$ corresponds to ideal undamped motion.

```python
A = 1.0
omega = 2.0
phi = 0

t = np.linspace(0, 20, 2000)

gamma_values = [0.0, 0.1, 0.3, 0.6]

for gamma in gamma_values:
    x_damped = A * np.exp(-gamma * t) * np.cos(omega * t + phi)
    plt.plot(t, x_damped, label=f"γ = {gamma}")

plt.title("Effect of Damping on Simple Harmonic Motion")
plt.xlabel("Time (s)")
plt.ylabel("Position")
plt.legend()
plt.show()
```

### Interpretation

As the damping coefficient $\gamma$ increases, the oscillations lose amplitude more quickly. When $\gamma = 0$, the oscillator behaves like an ideal simple harmonic oscillator with constant amplitude. For larger damping values, the exponential factor $e^{-\gamma t}$ causes the motion to decay toward zero more rapidly.

This shows how damping removes energy from the system over time, making the oscillation less persistent.

## 10. Driven Oscillations and Resonance

A driven oscillator is a system that is acted on by an external periodic force. In many physical systems, the oscillator responds most strongly when the driving frequency is close to the natural frequency of the system.

This effect is called resonance.

For a simplified driven oscillator, the steady-state amplitude depends on the driving frequency. The amplitude response can be modeled as:

$$
A(\omega_d) = \frac{F_0}{\sqrt{(k - m\omega_d^2)^2 + (b\omega_d)^2}}
$$

where:
- $F_0$ is the driving force amplitude
- $m$ is the mass
- $k$ is the spring constant
- $b$ is the damping coefficient
- $\omega_d$ is the driving angular frequency

The natural angular frequency is:

$$
\omega_0 = \sqrt{\frac{k}{m}}
$$

Resonance occurs when the driving frequency is near the natural frequency.

```python
m = 1.0
k = 4.0
b = 0.3
F0 = 1.0

omega_0 = np.sqrt(k / m)

omega_d_values = np.linspace(0.1, 5, 1000)

amplitude_response = F0 / np.sqrt(
    (k - m * omega_d_values**2)**2 + (b * omega_d_values)**2
)

plt.plot(omega_d_values, amplitude_response)
plt.axvline(omega_0, linestyle="--", label=f"Natural frequency = {omega_0:.2f}")

plt.title("Resonance Curve for a Driven Oscillator")
plt.xlabel("Driving Angular Frequency")
plt.ylabel("Steady-State Amplitude")
plt.legend()
plt.show()
```

### Interpretation

The resonance curve shows that the oscillator has the largest amplitude response when the driving angular frequency is close to the natural angular frequency of the system. For this example, the natural frequency is $\omega_0 = 2.0$.

At low or high driving frequencies, the response amplitude is smaller. Near resonance, the external driving force transfers energy efficiently into the oscillator, producing a much larger response.

Damping prevents the amplitude from becoming infinitely large and broadens the resonance peak.

## 11. Effect of Damping on Resonance

Damping affects both the height and width of the resonance peak. Lower damping produces a sharper, taller resonance peak, while higher damping reduces the maximum amplitude and broadens the response.

```python
m = 1.0
k = 4.0
F0 = 1.0

omega_0 = np.sqrt(k / m)
omega_d_values = np.linspace(0.1, 5, 1000)

b_values = [0.1, 0.3, 0.8, 1.5]

for b in b_values:
    amplitude_response = F0 / np.sqrt(
        (k - m * omega_d_values**2)**2 + (b * omega_d_values)**2
    )
    plt.plot(omega_d_values, amplitude_response, label=f"b = {b}")

plt.axvline(omega_0, linestyle="--", label=f"Natural frequency = {omega_0:.2f}")

plt.title("Effect of Damping on Resonance")
plt.xlabel("Driving Angular Frequency")
plt.ylabel("Steady-State Amplitude")
plt.legend()
plt.show()
```

### Interpretation

As damping increases, the resonance peak becomes lower and broader. With low damping, the oscillator responds very strongly near the natural frequency, producing a sharp resonance peak. With higher damping, energy is dissipated more quickly, reducing the maximum response.

This shows that damping controls how strongly a system resonates when driven near its natural frequency.
